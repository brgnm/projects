```python
df.drop(columns=['review_date'], inplace=True)

df['review_rating'].fillna(df['review_rating'].mean(), inplace=True)

df['work_hours'].fillna('No Data', inplace=True)

df['rating'] = df['rating'].clip(lower=1, upper=5)
df['review_rating'] = df['review_rating'].clip(lower=1, upper=5)

df['review_text'] = df['review_text'].str.replace(r'[^a-zA-Zа-яА-ЯёЁіІңҢәӘөӨұҰқҚғҒүҮұҰ\s]', '', regex=True)
df['review_text'] = df['review_text'].str.lower()
df['review_text'].replace(r'^\s*$', 'No Review', regex=True, inplace=True)
df = df[~df['review_text'].str.isnumeric()]

df.drop_duplicates(subset=['id', 'review_text'], keep='first', inplace=True)
```

```python
import ast

def extract_work_hours_features(work_hours_str):
    try:
        work_hours_data = ast.literal_eval(work_hours_str)

        work_day_count = 0
        work_time_count_per_day = 0

        for day, data in work_hours_data.items():
            if 'working_hours' in data and len(data['working_hours']) > 0:
                for hours in data['working_hours']:
                    start_time = hours['from']
                    end_time = hours['to']

                    if start_time == '09:00' and end_time == '22:00':
                        work_time_count_per_day = 13
                        work_day_count += 1
                    else:
                        start_hour = int(start_time.split(':')[0])
                        end_hour = int(end_time.split(':')[0])
                        work_time_count_per_day = (end_hour - start_hour)
                        work_day_count += 1

        if work_day_count == 0:
            work_day_count = 7
            work_time_count_per_day = 24

        return work_day_count, work_time_count_per_day
    except (ValueError, TypeError, KeyError) as e:
        return 7, 24

df[['work_day_count', 'work_time_count_per_day']] = df['work_hours'].apply(
    lambda x: pd.Series(extract_work_hours_features(str(x)))
)

print(df[['work_day_count', 'work_time_count_per_day']].head())
```
